```python
# Loading Day 1 Telecom dataset

tel_day1 = pd.read_csv('/content/sample_data/Telcom_dataset1.csv')
tel_day1
```

```python
# Loading Day 2 Telecom dataset

tel_day2 = pd.read_csv('/content/sample_data/Telcom_dataset2.csv')
tel_day2
```

# Data Cleaning

```python
# Rename the column names in the 3 datasets to ensure they are all similar
# Rename column name in dataset 1 with error
tel_day1 = tel_day1.rename(index=str, columns={"PRODUTC": "PRODUCT","DATETIME": "DATE_TIME"})
tel_day1
```

```python
# Rename column name in dataset 2 with error
tel_day2 = tel_day2.rename(index=str, columns={"DW_A_NUMBER": "DW_A_NUMBER_INT","DW_B_NUMBER": "DW_B_NUMBER_INT"})
tel_day2
```

```python
# Rename column name in dataset 3 with error
tel_day3 = tel_day3.rename(index=str, columns={"CELLID": "CELL_ID","SIET_ID": "SITE_ID"})
tel_day3
```

# Merge the Datasets

```python
# Merge the 3 days datasets

three_days_data = pd.concat((tel_day1, tel_day2, tel_day3), axis=0, ignore_index=True)
three_days_data
```

```python
# Check if we have any duplicates

three_days_data.duplicated()

# Drop the duplicates
three_days_data.drop_duplicates()
```

```python
# Check for null values
three_days_data.isnull()

# Drop missing observations
df_no_missing = three_days_data.dropna()
df_no_missing

# # Drop rows where all cells in that row is NA
df_cleaned = three_days_data.dropna(how='all')
df_cleaned

# # Drop column if they only contain missing values
three_days_data.dropna(axis=1, how='all')

# # Fill in missing data with zeros
three_days_data.fillna(0)
```

```python
# Merge the telecom 3 days datasets with the cells-geo dataset
# Rename the site_code column in the cells_geo data to match the site_ID in the merged three days telecom data

geo_data = geo_data.rename(columns = {'SITE_CODE':'SITE_ID'})
geo_data
```

```python
# Merge the telecom 3 days datasets with the cells-geo dataset

merged_all_data = pd.concat((three_days_data, geo_data), axis=0, ignore_index=True)
merged_all_data
```

```python
# From the above, there are many null values

merged_all_data.isnull()

# Drop missing observations
no_missing = merged_all_data.dropna()
no_missing

# # # Drop rows where all cells in that row is NA
cleaned = merged_all_data.dropna(how='all')
cleaned

# # # Drop column if they only contain missing values
merged_all_data.dropna(axis=1, how='all')

# # # Fill in missing data with zeros
merged_all_data.fillna(0)
```

```python
#Changing the DATE_TIME column to datatype datetime
merged_all_data["DATE_TIME"]= pd.to_datetime(merged_all_data['DATE_TIME'])
merged_all_data

#Finding cities most used during business hours
business_hrs = merged_all_data[(merged_all_data['DATE_TIME'] > 2012,5,6,7,0,0) & (merged_all_data['DATE_TIME'] < 2012,5,8 17,0,0)]
business_hrs
```

```python
# Most used city for the three days?
#Most used city for the three days, we use nlargest() function

#The result prints cocody as the most used city for the three days

most_used_city = merged_all_data['VILLES'].value_counts(2)     

most_used_city.nlargest(2)
```
